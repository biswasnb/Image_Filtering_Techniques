```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path =r"/content/CV1.jpg"
img = cv2.imread(img_path)
```

Gaussian Blur

```python
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis('off')
    plt.show()

    gaussian_blur = cv2.GaussianBlur(img, (15, 15), 0)
    gaussian_blur_rgb = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)
    plt.imshow(gaussian_blur_rgb)
    plt.title("Gaussian Blurred Image")
    plt.axis('off')
    plt.show()
```

Gaussian Blur with grayscale

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale for filtering


gaussian_blur = cv2.GaussianBlur(image_gray, (5, 5), 0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Gaussian Blurred Image")
plt.imshow(gaussian_blur, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Median Blur

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
median_blur = cv2.medianBlur(img, 5)
median_blur_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Median Blurred Image")
plt.imshow(median_blur_rgb)
plt.axis('off')
plt.tight_layout()
plt.show()
```

Sobel Blur

```python
    sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
    sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)

    abs_grad_x = cv2.convertScaleAbs(sobelx)
    abs_grad_y = cv2.convertScaleAbs(sobely)
    grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Sobel Edge Detection")
    plt.imshow(grad, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#  Sobel filters
sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)


sobel_combined = cv2.magnitude(sobel_x, sobel_y)


sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Sobel X")
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Sobel Y")
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Sobel Combined")
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
